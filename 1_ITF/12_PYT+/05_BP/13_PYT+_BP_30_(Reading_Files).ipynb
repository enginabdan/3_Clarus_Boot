```python
# -----------------'mode' Parameter----------------
```

```python
# "r" : Open for reading (default). If the file doesn't exist, FileNotFoundError will raise.
```

```python
# "a" : Open for writing. It will append to the end of the file if it already exists. If there is no file, it will create it.
```

```python
# "w" : Open for writing. It will be overwritten if the file already exists. If there is no file, it will create it.
```

```python
# "x" : Open for exclusive creation, it will fail if the file already exists.
```

```python
# "b" : Open in binary mode.
```

```python
# "t" : Open as a text file (default).
```

```python
# "+" : Open for updating (reading and writing).
```

```python
# By default, open() function executes opening the file for reading ('r') as a text ('t').
# In other words, it defaults to 'r' or 'rt' which means open for reading in text mode.
```

```python
# Note that, as you can see above, depending on what you are going to do with the file, you can combine the modes according to your needs.
# If you want to open and be able to read, modify, and update the existing file you should set the mode as 'r+'.
```

```python
# Another important point we should cover about mode is that you can choose the format in which you want to open the file as text ('t') or binary ('b').
```

```python
# 'binary' mode return contents as bytes objects without any decoding.
# In 'text' mode, the contents of the file are returned as str
```

```python
# ----------------Reading the Files with '.read()' Method-------------------
```

```python
# You can use the following methods for reading files :

# .read(size)
# .readline(size)
# .readlines()
# using loops
```

```python
# ------.read(size)-----
# This method reads the size bytes (characters) of the file.
# If this parameter is not specified, the entire file is read at once.
```

```python
sea = open("fishes.txt", 'r')   
print(sea.read())  # displays the entire text content
sea.close()  # be sure to close the file
```

```python
sea = open("fishes.txt", 'r')   
print(sea.read(33))     # displays the first 33 chars of the text
print()
print(sea.read(25))     # displays the next 25 chars of the text
print()
sea.seek(0)             # changes the stream (cursor) position to zero
print(sea.read(33))     # displays the first 33 chars again
print()
print(sea.tell())       # returns the current stream (cursor) position
sea.close()
```

```python
# Note that; while .read(33) counting the first 33 bytes, it covers newline \n and space character of the file text.
# The offset= 0 indicating the first character of the text, '.seek(offset) method' changes the stream (cursor) position to a given offset.
```

```python
my_file = open("fishes.txt", 'r')   
print(my_file.read(5)) 
my_file.close()
```

```python
# --------Reading the Files with '.readline()' Method----------
```

```python
# -------.readline(size)----------------
# This method reads only size bytes from a single line of the text.
```

```python
sea = open("fishes.txt", 'r')   

print(sea.readline())  # displays the first line of the text
print(sea.readline())  # displays the second line
print(sea.readline())  # each time it goes to the new line

sea.close()
```

```python
# Note that; .readline() reads the entire line.
# Pay attention to the output that there are empty lines between each line.
```

```python
# The first line of the text contains 26 bytes (characters) except the newline char (\n).
# But, note that the newline character (\n) is considered as a part of each line.
# Then the first line of text consists of 27 characters.
```

```python
sea = open("fishes.txt", 'r')   

print(sea.readline(13))
print(sea.readline(13))
print(sea.readline(13))
print(sea.readline(13))

sea.close()
```

```python
# As you can see, the .readline(13) reads 13 chars through the line and it continues to read 13 chars each until the line ends, and only then passes to the next line.
# In this example, considering the first line consists of 27 characters, the print(sea.readline(13)) syntax can read the entire line in a total of three repetitions.
# The output has two empty lines.
# One of them is caused by the last character (\n) of the first line and the other is because of the default empty line that the .readline() method puts between each line.
```

```python
# -----------------------Reading the Files with '.readlines()' Method----------------
```

```python
# -------------------.readlines()----------------
```

```python
# reads the entire file line by line in the list form.
```

```python
sea = open("fishes.txt", 'r')   

print(sea.readlines())

sea.close()
```

```python
# If we make the file read using one of these methods, each time we use the print(method) syntax it continues reading from where the cursor is.
```

```python
sea = open("fishes.txt", 'r')   

print(sea.readline())  # first line
print(sea.readlines())  # the rest of the lines

sea.close()
```

```python
sea = open("fishes.txt", 'r')   

print(type(sea.readlines()))

sea.close()
```

```python
# Although it is very good to read relatively small files with this method, but when large files are involved, this method can be inefficient.
```

```python
# Reading the Files with Loops
```

```python
# ------------Using Loops-------------
```

```python
sea = open("fishes.txt", 'r')   

for line in sea:
    print(line)

sea.close()
```

```python
# especially large text files can be read easily and practically.
# This method gives us the opportunity to read the line we want, as well as freeing us from syntax repetition.
```

```python
# Since the type of sea.readlines() object is a list, we can use it as an iterator and read the whole file in the same way.
```

```python
sea = open("fishes.txt", 'r')   

for line in sea.readlines():
    print(line)

sea.close()
```

```python
# The Matter of Closing the Files
```

```python
# Do not leave the file open.
# Keep in mind that closing the file is paramount.
```

```python
with open("fishes.txt", "r") as sea :
    print(sea.read())  # needs indented code block
```

```python
# we don't need to use the close() method to close the file. It does this on its own.
```

```python
with open("fishes.txt", "r") as my_file :
    print(my_file.read())
```
