```python
try :
    statement
    ...
except :
    statement
    ...
else :
    statement
    ...
finally :
    statement
    ...
```

```python
# executes all statements between try and except
# no exception, then the try block is successfully executed and finished
# If an exception raises, the rest of the try block is ignored
# checks if the type of exception matches the exception specified
# executes the except block and resumes executing the program after the try-except block
# exception doesn’t match the exception named in the except syntax, execution of your program stops with a traceback error message.
# else-block is only executed if there raised no exceptions.
# finally syntax is always executed before leaving the try-except block, whether an exception was raised or not
```

```python
while True:
    no_one = int(input("The first number please : "))
    no_two = int(input("The second number please : "))
    try:
        division = no_one / no_two  # normal part of the program
    except ZeroDivisionError:
        print("You can't divide by zero! Try again.")  # executes when division by zero
    else:
        print("The result of the division is : ", division)  # executes if there is no exception
    finally:
        print("Thanks for using our mini divison calculator! Come again!")
        break  # exits the while loop
```

```python
# Errors
# 1. Syntax Errors
# 2. Runtime Errors
# 3. Logic Errors
```

```python
# Syntax Errors
# This code will not run at all
x = 42
y = 206
if x == y
    print("Success")
```

```python
# Runtime Errors
# This code will fail when run
x = 42
y = 0
print(x / y)
```

```python
# Catching Runtime Errors
x = 4
y = 0
try :
    print(x / y)
except ZeroDivisionError as e :
    # Optional, log e somewhere
    print("Sorry, something went wrong")
except :
    print("Something really went wrong")
finally :
    print("This always runs on success or failure")
```

```python
# Logic Errors
# This code won't run at all
x = 206
y = 42
if x < y :
    print(str(x) + "is greater than" + str(y))
```
