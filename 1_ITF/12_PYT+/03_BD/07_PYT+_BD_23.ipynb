```python
with open("istiklal.txt", "a", encoding="utf-8") as engin :
    engin.write("\n")

    for i in new_line :
        if i == "\n" :
            i.replace("\n", "")  # pass de yazabiliriz
        else :
            engin.write(i)

with open("istiklal.txt", "r", encoding="utf-8") as engin :
    print(engin.read())
```

```python
veri = ["a", "b", True, (False, 1), {"1" : 2}, [1, 2], {"2" : "two"}, "c", 23, 0]

dict_typ = {}

for n in veri :
    keys = dict_typ.keys()
    if type(n) in keys :
        dict_typ[type(n)] += 1
    else :
        dict_typ[type(n)] = 1

print(dict_typ)
```

```python
# ----------------------{}.fromkeys(keys, 0)-------------------------
```

```python
veri = ["a", "b", True, (False, 1), {"1" : 2}, [1, 2], {"2" : "two"}, "c", 23, 0]

tipler = ["int", "str", "bool", "list", "tuple", "dict", "set"]

{}.fromkeys(tipler, 0)
```

```python
toplam = {}.fromkeys(tipler, 0)
```

```python
for i in range(len(veri)) :
    if type(veri[i]) == int : toplam["int"] += 1
    elif type(veri[i]) == str : toplam["str"] += 1
    elif type(veri[i]) == bool : toplam["bool"] += 1
    elif type(veri[i]) == list : toplam["list"] += 1
    elif type(veri[i]) == tuple : toplam["tuple"] += 1
    elif type(veri[i]) == dict : toplam["dict"] += 1
    elif type(veri[i]) == set : toplam["set"] += 1
toplam
```

```python
def my_factorial(n) :
    if n == 0 : return 1
    else :
        return n * my_factorial(n-1)
```
