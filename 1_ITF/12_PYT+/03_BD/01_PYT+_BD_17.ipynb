```python
def calc(a, b, c) :
    if c == "+" :
        return(a + b)
    elif c == "-" :
        return(a - b)
    elif c == "*" :
        return(a * b)
    elif c == "/" :
        return(a / b)
    elif c == "%" :
        return(a % b)
    elif c == "//" :
        return(a // b)
    else :
        return("Enter a valid character!!!")
calc(8, 3, "%")
```

```python
# ARMSTRONG NUMBER
while True :
    number = input("Enter a positive number.")
    digits = len(number)
    summ = 0
    if not number.isdigit() :
        print("Enter a valid number!")
    elif int(number) >= 0 :
        for i in range(digits) :
            summ = summ + int(number[i]) ** digits
        if summ == int(number) :
            print(number, "is an Armstrog number.")
            break
        else :
            print(number, "is not an Armstrog number.")
            break
```
