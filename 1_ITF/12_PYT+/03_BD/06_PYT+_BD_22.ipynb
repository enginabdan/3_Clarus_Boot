```python
fruits = ["banana", "mango", "pear", "apple", "kiwi", "grape"]
while True :
    try :
        index = int(input("Enter an index!"))
        print(f"My favorouite fruit is : {fruits[index]}")
        break
    except IndexError :
        print("Out of Index number, please try again :")
    except ValueError :
        print("Enter a valid integer number!")
```

```python
fruits = ["banana", "mango", "pear", "apple", "kiwi", "grape"]
counter = 3
while counter > 0 :
    try :
        index = int(input("Enter an index!"))
        print(f"My favorouite fruit is : {fruits[index]}")
    except (IndexError, ValueError, NameError, TypeError) :
        counter -= 1
        print(f"You have {counter} chances left!")
        print("Please try again :")
    else :
        print("Congrats! You have entered a valid input.")
        break
    finally :
        print("Our fruits are always fresh!")
```

```python
fruits = ["banana", "mango", "pear", "apple", "kiwi", "grape"]
counter = 3
while counter > 0 :
    try :
        index = int(input("Enter an index!"))
        print(f"My favorouite fruit is : {fruits[index]}")
    except (IndexError) :
        counter -= 1
        print(f"You have {counter} chances left!")
        print("Please try again, IndexError occured :")
    except (ValueError) :
        counter -= 1
        print(f"You have {counter} chances left!")
        print("Please try again, ValueError occured :")
    except (NameError) :
        counter -= 1
        print(f"You have {counter} chances left!")
        print("Please try again, NameError occured :")
    except (TypeError) :
        counter -= 1
        print(f"You have {counter} chances left!")
        print("Please try again, TypeError occured :")
    else :
        print("Congrats! You have entered a valid input.")
        break
    finally :
        print("Our fruits are always fresh!")
```

```python
meyveler = ["banana", "mango", "pear", "apple", "kiwi", "grape"]
with open("fruits.txt", "w", encoding="utf-8") as fruits :
    for i in meyveler :
        fruits.write(i + "\n")

with open("fruits.txt", "r", encoding="utf-8") as fruits :
    print(fruits.read())

with open("fruits.txt", "r", encoding="utf-8") as fruits :
    print(fruits.readlines())
```
