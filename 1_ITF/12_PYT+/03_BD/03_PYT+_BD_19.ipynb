```python
# BİRDEN FAZLA PARAMETREYİ **keyword ARGUMENT İLE ÇAĞIRMA
```

```python
def önceki(**parameter) :
    for x, y in parameter.items() :
        print(x, y)

önceki(a = "engin", b = "azra", c = "neda", d = "aden")
```

```python
def sonraki(a = "engin", b = "azra", c = "neda", d = "aden") :
    print(a, ": a",)
    print(b, ": b",)
    print(c, ": c",)
    print(d, ": d",)

x = {"a" : "ahmet", "b" : "mehmet", "c" : "kerem", "d" : "selman"}
sonraki(**x)
```

```python
def çiftler(gelinler, damatlar) :
    çiftler_listesi = []
    for x in zip(gelinler, damatlar) :
        çiftler_listesi.append(x)
    return çiftler_listesi

a = {"gelinler" : ["azra", "merve", "gülten"],
     "damatlar" : ["engin", "hakan", "nusret"]}
çiftler(**a)
```

```python
# LİST COMPREHENSİONLA TEK SATIRDA YAPMA
def çiftler(gelinler, damatlar) :
    return [x for x in zip(gelinler, damatlar)]

a = {"gelinler" : ["azra", "merve", "gülten"],
     "damatlar" : ["engin", "hakan", "nusret"]}
çiftler(**a)
```

```python
def is_valid(s) :
    while "()" in s or "[]" in s or "{}" in s :
        s = s.replace("()", "").replace("[]", "").replace("{}", "")
    return s == ""
```

```python
s = "([{}[]])"
is_valid(s)
```
