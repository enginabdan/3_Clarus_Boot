```python
# try :
#     statement
#     ...
# except :
#     statement
#     ...
# else :
#     statement
#     ...
# finally :
#     statement
#     ...

# Except çalışırsa : except + finally çalışır.
# Except çalışmazsa : try + else + finally çalışır.
```

```python
try :
    print("4" + 4)
except TypeError:
    print("Exception: There is TypeError!")
else :
    print("Exception not occured!")
finally :
    print("Finally: My turn")
```

```python
try :
    print("4" + 4)
except ValueError:
    print("TypeError var!")
else :
    print("Exception oluşmamış!")
finally :
    print("Bana geldi")
```

```python
while True :
    a = int(input("First :"))
    b = int(input("Second :"))
    try :
        div = a / b
    except ZeroDivisionError :
        print("Can not divide by zero!")
    else :
        print("Result is :", div)
    finally :
        print("Thnaks using, try again later!")
        break
```
