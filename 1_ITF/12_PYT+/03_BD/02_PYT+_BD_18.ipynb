```python
# Calculator problem
def calc(a, b, c) :
    if c == "-" :
        return(a - b)
    elif c == "+" :
        return(a + b)
    elif c == "*" :
        return(a * b)
    elif c == "/" :
        return(a / b)
    else :
        return("Enter a valid character!")
print(calc(3, 5, "*"))
```

```python
# Birden fazla parameter define edip sonrasında:
# Birden fazla Positional Argument veya:
# Birden fazla Keyword Argument çağırabiliriz
# def funct(*parameter) for positional arguments;
# def funct(**parameter) for keyword arguments.
```

```python
def fruiter(fruit1, fruit2) :
    print("I want to get", fruit1, "and", fruit2)

fruiter("orange", "banana")
```

```python
def fruiter(*fruit) :
    print("I want to get :")
    for i in fruit :
        print("-", i)

fruiter("orange", "banana", "melon", "ananas")
```

```python
def fruiter(*fruit) :
    print("I want to get:", end=" ")
    for i in fruit :
        print(i, ",", end=" ")

fruiter("orange", "banana", "melon", "ananas")
```

```python
# LOOP İLE YAPIM
def slicer(*numbers) :
    even_list = []
    odd_list = []
    for i in numbers :
        if i%2 == 0 :
            even_list.append(i)
        else :
            odd_list.append(i)

    print("Even list . ", even_list)
    print("Odd list . ", odd_list)
```

```python
slicer(1,2,3,4,5,6,7,8,9)
```

```python
# LIST COMPREHENSION İLE YAPIM
def slicer(*numbers) :
    print("even_list :", [i for i in numbers if i%2 == 0])
    print("odd_list :", [i for i in numbers if not i%2 == 0])

slicer(1,2,3,4,5,6,7,8,9)
```

```python
# Keyword Argument ile birden fazla parametre ve argument tanımlama çağırma
```

```python
def animals(**kwargs) :
    for key, value in kwargs.items() :
        print(value, "are", key)

animals(Carnivores="Lions", Omnivores="Bears", Herbivores="Deers", Nomnivores="Human")
```

```python
sözlük = dict(a="1", b="2", c="3", d="4")
sözlük
```

```python
def sözlük(**x) :
    for key, value in x.items() :
        print(key, value)

sözlük(a="1", b="2", c="3", d="4")
```

```python
def organizer(**x) :
    list_names = []
    list_ages = []
    for key, value in x.items() :
        list_names.append(key)
        list_ages.append(value)
    print(list_names, list_ages, sep="\n")

organizer(engin=39, azra=40, neda=13, aden=4)
```
