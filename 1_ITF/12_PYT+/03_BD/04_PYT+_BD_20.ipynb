```python
a = [9, 6, 7, 4]
b = [3, 6, 5, 8]
abs = map(lambda x, y : (x+y)/2, a, b)
print(*abs)
```

```python
a = ["you", "much", "hard"]
b = ["I", "You", "He"]
c = ["love", "ate", "works"]
concat = map(lambda x, y, z : x+" "+y+" "+z, b, c, a)
print(*concat)
```

```python
# -----------------FILTER-----------------
# ---------filter(function, sequence)-----
# iterablelara tek tek uygulari Trueları geçirip yazdırır
```

```python
a = [1,2,3,4,5,6,7,8,9]
even = filter(lambda x : x%2 == 0, a)
odd = filter(lambda x : x%2 != 0, a)
print(type(even))
print(list(even))
print(type(odd))
print(list(odd))
```

```python
a = ["apple", "swim", "clock", "me", "kiwi", "banana"]
b = filter(lambda x : len(x) < 5, a)
c = filter(lambda x : len(x) >= 5, a)
print(type(b))
print(list(b))
print(type(c))
print(list(c))
```

```python
a = ["apple", "swim", "clock", "me", "kiwi", "banana"]
b = filter(lambda x : len(x) < 5, a)
for i in b :
    print(i)
```

```python
k = ["a", "e", "i", "o", "u"]
l = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]
m = filter(lambda x : True if x in k else False, l)
print(*m)
```

```python
k = ["a", "e", "i", "o", "u"]
l = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]
m = filter(lambda x : x in k, l)
print(*m)
```
