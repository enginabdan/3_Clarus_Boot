Problem :

    1. Define a function to take a word and return negative meaning.
    2. Given a word, return a new word where "not " has been added to the front.
    3. However, if the word already begins with "not", return the string unchanged.

```python
def not_string(word):
  if len(word) >= 3 and word[:3] == "not":
      return word
  else:
      return "not " + word
```
