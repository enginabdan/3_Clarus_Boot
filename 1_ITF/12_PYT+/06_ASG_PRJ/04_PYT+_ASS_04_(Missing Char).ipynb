Problem :

    1. Given a non-empty string and an int n,
    2. return a new string where the character at index n has been removed.
    3. The value of n will be a valid index of a character in the original string
    4. (i.e. n will be in the range 0....len(str)-1 inclusive).

```python
def missing_char(word,n):
    front = word[:n]
    back = word[n+1:]
    return front + back
print(missing_char('kitchen', 4))
```
