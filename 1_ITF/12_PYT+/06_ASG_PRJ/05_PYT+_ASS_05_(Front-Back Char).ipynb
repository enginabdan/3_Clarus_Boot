Problem :

    1. Given a string,
    2. return a new string where the first and last chars have been exchanged.

```python
def front_back(word):
    if len(word) >= 2:
        return word[-1:] + word[1:-1] + word[:1]
    else:
        return word  

print(front_back('abcd'))
print(front_back('12345'))
print(front_back('a'))
```
