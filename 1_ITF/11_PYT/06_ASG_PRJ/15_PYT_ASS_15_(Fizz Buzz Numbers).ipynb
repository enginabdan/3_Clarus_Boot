Task : Print the Fizz Buzz numbers.

    1. Fizz Buzz is a famous code challenge used in interviews to test basic programming skills.
    2. It's time to write your own implementation.
    3. Print numbers from 1 to 100 inclusively following these instructions:
        a. if a number is multiple of 3, print "Fizz" instead of this number,
        b. if a number is multiple of 5, print "Buzz" instead of this number,
        c. for numbers that are multiples of both 3 and 5, print "FizzBuzz",
        d. print the rest of the numbers unchanged.
    4. Output each value on a separate line.

```python
for i in range(1,101):
  if i%15==0:
    print("FizzBuzz")
  elif i%5==0:
    print("Buzz")
  elif i%3==0:
    print("Fizz")
  else:
    print(i)
```
