Task: Count the number of each letter in a sentence.

1. The department you work for undertook a project construction that makes word / text analysis.

2. You are asked to calculate the number of letters or any chars in the sentences entered under this project.

3. Write a Python program that;
    
    a. takes a sentence from the user,
    
    b. counts the number of each letter of the sentence,
    
    c. collects the letters/chars as a key and the counted numbers as a value in a dictionary.

```python
word=input()
dictnry={}
for i in word:
    if i not in dictnry:
        dictnry[i]=1
    elif i in dictnry:
        dictnry[i]+=1
print(dictnry)
```
