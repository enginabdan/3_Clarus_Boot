Task: Find out if a given year is a "leap" year.

1. In the Gregorian calendar, three criteria must be taken into account to identify leap years:
    
    a. The year must be evenly divisible by 4;
    
    b. If the year can also be evenly divided by 100, it is not a leap year; unless...
    
    c. The year is also evenly divisible by 400. Then it is a leap year.
    
2. According to these rules,
    
    a. the years 2000 and 2400 are leap years,
    
    b. while 1800, 1900, 2100, 2200, 2300, and 2500 are not the leap years.
    
3. Write a Python program that prints such as "2020 is a leap year" if the given year by the user is a leap year, prints such as "2019 is not a leap year" otherwise.

```python
year=int(input("Enter a year!"))
if year%400==0:
  print(f"{year} is a leap year")
elif year%100==0:
  print(f"{year} is not a leap year")
elif year%4==0:
  print(f"{year} is a leap year")
else:
  print(f"{year} is not a leap year")
```
