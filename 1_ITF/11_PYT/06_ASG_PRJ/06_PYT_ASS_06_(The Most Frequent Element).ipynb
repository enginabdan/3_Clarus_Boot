Task : Find out the most frequent number and its frequency.

Write a program that;

    1. Finds out the most frequent number in the given list.
    
    2. Calculates its frequency.
    
    3. Prints out the result such as :

```python
numbers = [1, 3, 7, 4, 3, 0, 3, 6, 3]
tup_numbers = tuple(numbers)
result = max(set(numbers), key = numbers.count)
max_num = tup_numbers.count(result)
print(f"the most frequent number is {result} and it was {max_num} times repeated")
```
