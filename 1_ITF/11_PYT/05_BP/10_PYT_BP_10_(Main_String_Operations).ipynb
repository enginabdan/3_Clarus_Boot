```python
# ----------------Searching a String-----------------
# variable.endswith("",start, stop)
# variable.startswith("", start, stop)
# True---False
```

```python
text = 'www.clarusway.com'
print(text.endswith('.com'))
print(text.startswith('http:'))
```

```python
text = "www.clarusway.com"
print(text.startswith("w"))
print(text.endswith("om"))
```

```python
email = "clarusway@clarusway.com is my e-mail address"
print(email.startswith("@", 9))
print(email.endswith("-", 10, 32))
print(email.endswith("-", 10, 31))
print(email.endswith("m", 10, 33))
```

```python
phrase = "myemailaddress@clarusway.com"

print(len(phrase))
print(phrase.startswith("@", 14))
print(phrase.endswith(".", 15, 24))
```

```python
a = "www.cumhuriyet.com.tr"
print(a.startswith("i", 10))
print(a.endswith(".",2,4))
print(a.endswith(".",10,14))
print(a.endswith(".",10,15))
print(a.endswith(".",15,19))
```

```python
# ----------------Changing a String-----------------
```

```python
# Syntaxes
# 1. str.replace(old, new[, count])
# 2. str.swapcase()
# 3. str.capitalize()
# 4. str.upper()
# 5. str.lower()
# 6. str.title()
```

```python
a = "enginabdan"
b = a.replace("i", "+")
print(b)
```

```python
a = "engin abdan"
b = a.replace(" ", "+")
print(b)
```

```python
a = "engin abdan"
b = a.capitalize().replace("abdan", "ABDAN")
print(b)
```

```python
a = "engin abdan"
b = a.capitalize()
print(b)
```

```python
a = "engin abdan"
b = a.upper()
print(b)
```

```python
a = "ENGİN ABDAN"
b = a.lower()
print(b)
```

```python
a = "eNGİN aBDAN"
b = a.swapcase()
print(b)
```

```python
sentence = "I live and work in Virginia"

print(sentence.upper())  

print(sentence.lower())

print(sentence.swapcase())

print(sentence)  # note that, source text is unchanged
```

```python
sentence = "I live and work in Virginia"
title_sentence = sentence.title() 
print(title_sentence)  
changed_sentence = sentence.replace("i", "+")
print(changed_sentence)  
print(sentence)  # note that, again source text is unchanged
```

```python
sentence = "I live and work in Virginia"
swap_case = sentence.swapcase()
print(swap_case)
print(swap_case.capitalize())  # changes 'i' to uppercase and
# the rest to lowercase
```

```python
print("Actions speaks louder than words".upper().swapcase().capitalize())
```

```python
# ----------------Editing a String----------------- # str.strip()  # str.rstrip()  # str.lstrip()
```

```python
space_string = "     listen first      "
print(space_string.strip())  # removes all spaces from both sides
```

```python
source_string = "interoperability"
print(source_string.strip("yi"))  # removes trailing "y" or "i" or "yi" or "iy" from both sides
```

```python
source_string = "interoperability"
print(source_string.lstrip("in"))  # removes "i" or "n" or "in" or "ni" from the left side
```

```python
space_string = "     listen first      "
print(space_string.rstrip())  # removes spaces from the right side
```

```python
source_string = "interoperability"
print(source_string.rstrip("yt"))  # removes "y" or "t" or "yt" or "ty" from the right side
```
