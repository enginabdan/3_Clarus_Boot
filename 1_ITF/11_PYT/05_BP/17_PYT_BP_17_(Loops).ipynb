```python
print ("Let's play the guessing game!")
answer = 44
question = 'What number am I thinking of?  '

while True:
    guess = int(input(question))

    if guess < answer:
        print('Little higher')
    elif guess > answer:
        print('Little lower')
    else:  # guess == answer
        print('Are you a MINDREADER!!!')
        break
```

```python
a = ["one","two","three","four","five"]  # List
b = [1,2,3,4,5]
for x,y in zip(a, b):
  print(x, ":", y)
```

```python
a = ("one","two","three","four","five") # Tuple
b = (1,2,3,4,5)
for x, y in zip(a, b):
  print(x, ":", y)
```

```python
a = {"one","two","three","four","five"} # Set
b = {1,2,3,4,5}
for x, y in zip(a, b):
  print(x, ":", y)
```

```python
a = {"a":"1", "b":"2", "c":"3", "d":"4"}
b = a.items()
c = a.keys()
d = a.values()
print(b, c, d, sep="\n")
for x, y in zip(c, d):
  print(x, ":", y)
```
