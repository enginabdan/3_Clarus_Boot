```python
# The if statements check the condition.
# The condition is always a Boolean expresion.
# Its value equals either True or False.
```

```python
if True:
    print('it is true')
```

```python
empty_seat = 14

if empty_seat > 3:  # in this case, 14>3=True, so the body will execute
    print('there is still seat to sit')
```

```python
# comparison operators :  
# NOT EQUAL:              !=
# EQUAL:                  ==
# LESS THAN:              <
# GREATER THAN:           >
# GREATER THAN OR EQUAL:  >=  
# LESS THAN OR EQUAL:     <=
# Do not use assignment operator = in comparison statements.
# In comparisons, you have to use == operator for equality.
```

```python
x = 6
y = 9
print ("is x equal to y?                 :" , x == y)
print ("is x not equal to y?             :" , x != y)
print ("is x less than y?                :" , x < y)
print ("is x greater than y?             :" , x > y)
print ("is x less than or equal to y?    :" , x <= y)
print ("is x greater than or equal to y? :" , x >= y)
```

```python
# if condition1:  
#     execute body1
# else : # if condition1 is not ensured execute body2
#     execute body2

# Note that, else doesn't require any condition and
# the body2 requires 4-space indentation here.
```

```python
course = 'clarusway'

if course == "clarusway":
    print("you guaranteed the job")
else:
    print("think about it again")
```

```python
number = 5
if number <= 3:    
    print("Number is smaller than or equal to 3") 
else:  # Optional clause (you can only have one else)
    print("Number is bigger than 3")
```

```python
basket = ['apple', 'peach', 'blackberry']
fruit1 = "apple"
fruit2 = "banana"
if fruit1 not in basket:
  print("I can have a little")
else:
  print("I have already")
if fruit2 not in basket:
  print("I can have a little")
else:
  print("I have already")
```

```python
weight = 80

if weight > 100:
    print("That's too heavy!")
elif weight > 85:
    print("I can lift that!")       # We can add as many elifs as we need.
elif weight > 50:
    print("Less than I can lift")
else:
    print("That's too light!")
```
