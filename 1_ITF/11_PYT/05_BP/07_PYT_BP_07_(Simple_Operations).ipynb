```python
#--------in-place operation--------
# -= decrements the variable in place,                  a = a - number
# += increment the variable in place,                   a = a + number
# *= multiply the variable in place,                    a = a * number
# /= divide the variable in place,                      a = a / number
# //= ﬂoor divide the variable in place,                a = a // number
# %= returns the modulus of the variable in place,      a = a % number
# **= raise to power in place.                          a = a ** number
```

```python
x = 5
print ('value of x       : ', x)

x += 2
print ("2 more of x      : ", x, "\n") # using string expression '\n', 
                                       # we produce extra line. 
                                       # So that we had empty line.
y = 10
print ('value of y       : ', y)

y -= 2
print ("2 minus y        : ", y, "\n")

z = 6
print ('value of z       : ', z)

z *= 2
print ("2 times z        : ", z, "\n")
```
