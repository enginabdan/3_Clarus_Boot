```python
# The same as List, list[] but tuple()
# Tuple = Immutable, therefore can not change the values
# Can not use append, insert, del, pop ...
```

```python
empty_tuple = ()
```

```python
empty_tuple = ()
print(type(empty_tuple))
```

```python
try_tuple = ('love')
print(try_tuple)
print(type(try_tuple)) # it's not tuple type.
```

```python
try_tuple = ('love',)
print(try_tuple)
print(type(try_tuple)) # it's a tuple type, because of ",".
```

```python
empty_tuple_1 = tuple()
print(empty_tuple_1)
print(type(empty_tuple_1))
```

```python
a = tuple()
print(a)
print(type(a))
```

```python
a = (1,2,3,4,5)
b = list(a)
print(a)
print(type(a))
print(b)
print(type(b))
```

```python
a = [1,2,3,4,5]
b = tuple(a)
print(a)
print(type(a))
print(b)
print(type(b))
```

```python
my_tuple=(1, 4, 3, 4, 5, 6, 7, 4)
my_list = list(my_tuple)
print(type(my_tuple), my_tuple, sep="\n")
print(type(my_list), my_list, sep="\n")
```

```python
my_list = [1, 4, 3, 4, 5, 6, 7, 4]
my_tuple = tuple(my_list)
print(type(my_list), my_list, sep = "\n")
print(type(my_tuple), my_tuple, sep = "\n")
```

```python
a = "engin"
b = tuple(a)
print(b)
```

```python
a = "engin"
b = list(a)
print(b)
```
