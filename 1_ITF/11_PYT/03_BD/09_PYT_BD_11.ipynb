```python
text = "{:.2f}, {:.3f}, {:.4f}".format(3.1463, 5.3676, 7.324567)
print(text)   # yuvarlayarak alıyor noktadan sonra
```

```python
text = "{:.2s} - {:.3s} - {:.4s}".format("3.1463", "5.3676", "7.324567")
print(text)  # stringden alınacak karakter miktarı
```

```python
text = "{:>10}".format("engin")
print(text)  # toplam 10 karakterde sağa yasladı
```

```python
text = "{:<10}".format("engin")
print(text)  # toplam 10 karakterde sola yasladı
```

```python
text = "azra {:>10} azra".format("engin")
print(text)  # ortada toplam 10 karakterde sağa yasladı
```

```python
text = "azra {:<10} azra".format("engin")
print(text)  # ortada toplam 10 karakterde sola yasladı
```

```python
text = "{:<10} azra".format("engin")
print(text)  # solda toplam 10 karakterde sola yasladı
```

```python
text = "{:>10} azra".format("engin")
print(text)  # solda toplam 10 karakterde sağa yasladı
```

```python
text = "{:^10}".format("engin")
print(text)  # 10 karakterde ortada yasladı
```

```python
text = "azra {:^10} azra".format("engin")
print(text)  # ortada toplam 10 karakterde ortaya yasladı
```
