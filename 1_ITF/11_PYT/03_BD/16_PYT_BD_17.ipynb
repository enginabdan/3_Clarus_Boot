```python
a = [11, 2, 24, 61, 48, 33, 3]
evens = []
odds = []
for x in a :
    if x%2 == 0 :
        evens.append(x)
    else :
        odds.append(x)
print("The number of even numbers: ", len(evens))
print("The number of odd numbers: ", len(odds))
```

```python
a = [11, 2, 24, 61, 48, 33, 3]
evens = 0
odds = 0
for x in a :
    if x%2 == 0 :
        evens += 1
    else :
        odds += 1
print("The number of even numbers: ", evens)
print("The number of odd numbers: ", odds)
```

```python
# -------------------Nested for Loops------------------------
# for variable1 in iterable1 :
#     for variable2 in iterable2 :
#         body   
```

```python
a = ["I am ", "You are "]
b = ["happy", "confident"]
for i in a :
    for ii in b :
        print(i + ii)
```

```python
a = [1, 2, 3, 4, 5, 6]
b = []
c = []
d = []
e = []
for i in a :
    if i%2 == 0 :
        b.append(i)
        c.append(i**2)
    else :
        d.append(i)
        e.append(i**2)
print(b, c, d, e, sep="\n")
```
