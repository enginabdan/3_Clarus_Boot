```python
answer = 40
Question = "Reach the 2 digit number :"
print("Let's play!")
while True :
    guess = int(input("Question"))
    if guess < answer :
        print(guess)
        print("Little higher")
    elif guess > answer :
        print(guess)
        print("Little lower")
    else :
        print(guess)
        print("Congratulations!")
        break  # While larda if state kullanıp bilince break ile kapat.
```
