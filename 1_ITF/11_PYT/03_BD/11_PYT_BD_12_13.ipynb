```python
sorted(["malatya","bursa","izmir","ankara","kars"])
```

```python
a = ["malatya","bursa","izmir","ankara","kars"]
sorted(a)  # ASCII koda göre sıraladı
```

```python
a = ["malatya","bursa","izmir","ankara","kars"]
sorted(a, key = len) # uzunluğa göre sırala
```

```python
a = ["malatya","bursa","izmir","ankara","kars"]
sorted(a, key = len, reverse=True) # uzunluğa göre sırala
```

```python
a = ["malatya","bursa","izmir","ankara","kars"]
a.sort(key = len, reverse=True)
print(a)
```

```python
a = ["malatya","bursa","izmir","ankara","kars"]
a.sort(key = len, reverse=False)
print(a)
```
