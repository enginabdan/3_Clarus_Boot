```python
# "generosity wins in all circumstances."
phrase = "{2} {3} {1} {0}".\
format("circumstances.", "in all", "generosity", "wins")
print(phrase)
```

```python
# "generosity wins in all circumstances."
a = "circumstances"
b = "generosity"
phrase = "{b} {1} {0} {a}".\
format("in all", "wins", a = "circumstances", b = "generosity")
print(phrase)
```

```python
# "generosity wins in all circumstances."
a = "circumstances"
b = "generosity"
phrase = "{b} {} {} {a}".\
format("wins", "in all", a = "circumstances", b = "generosity")
print(phrase)
```
