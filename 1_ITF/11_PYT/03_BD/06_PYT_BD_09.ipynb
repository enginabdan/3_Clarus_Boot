```python
# Format içnde positionallar sonra keywordler yazılır. 
# "california" = positional argument
# state="california" = keyword argument
```

```python
print("{}, {x}, {}, {}".format("a", x = "b","c", "d"))
```

```python
print("{}, {x}, {}, {}".format("a", "c", "d", x = "b"))
```

```python
print("{y}, {x}, {}, {}".format(y = "a", "c", "d", x = "b"))
```

```python
print("{y}, {x}, {}, {}".format("c", "d", x = "b", y = "a"))
```

```python
print("{y}, {}, {x}, {}".format(1, "d", x = "b", y = "5"))
```

```python
print("{y}, {1}, {x}, {0}".format(1, "d", x = "b", y = "5"))
```

```python
print("{} is the most {} state of the {}"\
      .format("california", "crowded", "USA"))
# Positional
```

```python
print("{state} is the most {adjective} state of the {country}"\
      .format(state="california", country="USA", adjective="crowded"))
# keyword
```

```python
state = "California"
country = "USA"
adjective = "crowded"
print("{} is the most {} state of the {}"\
      .format(state, adjective, country))
# keyword
```

```python
state = "California"
country = "USA"
adjective = "crowded"
print("{0} is the most {2} state of the {1}"\
      .format(state, country, adjective))
# Keyword
```

```python
print("{} is the most {adjective} state of the {country}"\
      .format("california", country="USA", adjective="crowded"))
# positional + Keyword
```

```python
# Format içine önce positionallar,
# Sonra keywordler
# positionalları index no ile dağıtabilirsin
# string içi süslü paranteze positionalın indexini yazabilirsin
# keywordler sona ve indexleyemezsin, variable süslü içine yaz
```
