```python
# SUMMARY

#FUNCTIONAL
plt.figure(figsize=(4,4))
plt.plot(age, salary)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");

# OO METHOD
fig, ax = plt.subplots(2, 2, figsize=(8,8))
ax[0,1].plot(age,salary, 'r')
ax[0,1].set_xlabel("age")
ax[0,1].set_ylabel('salary')
ax[0,1].set_title("Salary by Age");

# ADD_AXES METHOD
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y)
ax2 = fig.add_axes([.25,.5,.5,.5])
ax2.plot(x)
ax3 = fig.add_axes([.1,.1,.2,.2])
ax3.plot(x)

# LW LS
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y, color="red", lw=2, ls="-.")
ax.plot(x,y-1, color="blue", linewidth=4, ls="--")
ax.plot(x,y-2, color="yellow", lw=6, ls=":")
ax.plot(x,y-3, color="black", linewidth=8, ls="--");

# MARKER
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y, marker="+", markersize=4)
ax.plot(x,y-1, marker="o", markersize=4)
ax.plot(x,y-2, marker="s", markersize=4)
ax.plot(x,y-3, marker="1", markersize=4);

# LOCATION OF LABEL
plt.plot(age, salary, label="salary")
plt.plot(age, salary_2, label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age")
plt.legend(loc=4);

# XLIM YLIM AXVLINE AXHLINE
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
ax[0].plot(x, x**2, x,x**3, x,x**4)
ax[0].set_xlim([2,8])
ax[0].set_ylim([20,1000])
ax[0].axvline(x=3, ls="--")
ax[1].plot(x, x**2, x,x**3, x,x**4)
ax[1].set_xlim([2,4])
ax[1].set_ylim([9,100])
ax[1].axhline(y=60, ls="--")
plt.tight_layout();
```
