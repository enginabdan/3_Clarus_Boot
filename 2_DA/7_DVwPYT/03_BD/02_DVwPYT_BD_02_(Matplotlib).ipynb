```python
# SUMMARY

# LABEL LEGEND
plt.plot(age, salary,label="1")
plt.plot(age, salary_2, label="2")
plt.legend();

# MULTI GRAPH WITH FOR LOOP
def multi_graf(rows,cols):
    fig, ax = plt.subplots(nrows=rows, ncols=cols)
    for i in range(rows):
        for j in range(cols):
            ax[i,j].plot(age,salary)
            plt.tight_layout()
multi_graf(2,3)

# SEABORN DATASETS
sns.get_dataset_names()
tips = sns.load_dataset("tips")

# SCATTER PLOT
plt.scatter(tips.tip, tips.total_bill);

# BAR PLOT
day = tips.groupby("day")["total_bill"].sum().index
tip = tips.groupby("day")["tip"].sum()
total = tips.groupby("day")["total_bill"].sum()
fig, ax = plt.subplots()
ax.bar(day, total, label="total")
ax.bar(day, tip, label="tip")
plt.legend();

# HISTOGRAM PLOT
plt.hist(tips.total_bill,bins=50)
    # Right skewness
    # -0,5 // +0,5  Normal Distrubition
    # +0,5 // +1,0  Moderate Distrubition
    # > +1,0        Skewed
tips.total_bill.skew() 
    # Skewed >>> skewness > +1,0
    # Right Skewed >>> mean is on the right side of median

# BOX PLOT
plt.figure(figsize=(10,4))
plt.boxplot([tips.total_bill, tips.tip], vert=False)
plt.tight_layout();
```

```python
def multi_graf(rows,cols):
    fig, ax = plt.subplots(nrows=rows, ncols=cols)
    for i in range(rows):
        for j in range(cols):
            ax[i,j].plot(age,salary)
            plt.tight_layout()
multi_graf(2,3)
```
