```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# We use the categorical plots when the data set has categories.
# You can see the categories in the pictures above as weather condition
# (drizzle, rain, sun weather, etc) or days of a week (Thur, Fri, Sat, and Sun)

# In this lesson, you will discuss some plots that allow visualizing the categorical data.
# These plots are:
    # barplot,
    seaborn.barplot(x=None, y=None, hue=None, data=None, estimator=function, ci=95)
    # countplot,
    seaborn.countplot(x=None, y=None, hue=None, data=None, orient=None)
    # boxplot,
    seaborn.boxplot(x=None, y=None, hue=None, data=None, orient=None)
    # violinplot,
    seaborn.violinplot(x=None, y=None, hue=None, data=None, palette=None)
    # swarmplot,
    # boxenplot

# Barplot gives also aggragation result about the grouped column value / ci, var, std, max...
# Countplot does not give these info
```
