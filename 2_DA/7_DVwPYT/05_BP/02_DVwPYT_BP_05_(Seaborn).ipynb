```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

### COMPARISON PLOT

```python
sns.catplot(data=df, x="gender", y="math score", kind="box")
```

```python
sns.catplot(data=df, x="gender", y="math score", kind="box", row="lunch")
```

```python
sns.catplot(data=df, x="gender", y="math score", kind="box", hue="lunch")
```

```python
sns.catplot(data=df, x="gender", y="math score", kind="box", col="lunch")
```

```python
sns.catplot(data=df, x="gender", y="math score", kind="box", col="lunch", row="test preparation course")
```

```python
sns.catplot(data=df, x="gender", y="math score", kind="box", col="lunch", row="race/ethnicity")
```

```python
sns.catplot(data=df, x="gender", y="math score", kind="violin", col="lunch", row="race/ethnicity")
```

### PAIR GRID

```python
sns.pairplot(df)
```

```python
sns.PairGrid(df)
```

```python
g = sns.PairGrid(df)
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.kdeplot)
g = g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(df, hue="gender")
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.kdeplot)
g = g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(df, hue="gender")
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.kdeplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend()
```
