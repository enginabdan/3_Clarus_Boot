```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# In this lesson you will learn;
    # Subplot with the OO Method,
    # "Setting" the number of "rows" and "columns" to the subplots,
    # How to prevent overlapping subplots or figures with "tight_layout()"" method,
    # "Setting parameters" on subplots,
    # "Adding sub-titles" with  "suptitle method",

# Functional Method
    plt.subplot(2,1,1)
    plt.plot(x1, y1) 
    plt.subplot(2,1,2)
    plt.plot(x2, x2)

# Object-Oritented Method
    fig,axes=plt.subplots(1,2)
    # or  
    fig,axes=plt.subplots(nrows=1,ncols=2)
    axes[].plot(x ,y)


fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)
axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel("Y LABEL 1,0")
axes[1][0].set_xlim(2,6)

axes[1][1].plot(a,b)
axes[1][1].set_title("TITLE 1,1")

fig.suptitle("Figure Level", fontsize=16)
plt.tight_layout()
fig.set_figwidth(10)
fig.savefig("new_subplots.png", bbox_inches="tight")
```
