```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
sns.scatterplot(x="salary", y="sales", data=df)
```

```python
# figsize()
plt.figure(figsize=(12,4))
sns.scatterplot(x="salary", y="sales", data=df)
```

```python
# dpi=
plt.figure(figsize=(12,4), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df)
```

```python
# hue=
plt.figure(figsize=(12,4), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df, hue="level of education")
# Choosing a palette from Matplotlib's cmap: https://matplotlib.org/tutorials/colors/colormaps.html
```

```python
# hue=
plt.figure(figsize=(12,4), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df, hue="salary")
# Choosing a palette from Matplotlib's cmap: https://matplotlib.org/tutorials/colors/colormaps.html
```

```python
# hue=
plt.figure(figsize=(12,4), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df, hue="level of education")
# Choosing a palette from Matplotlib's cmap: https://matplotlib.org/tutorials/colors/colormaps.html
```

```python
# Palette=
# Choosing a palette from Matplotlib's cmap: https://matplotlib.org/tutorials/colors/colormaps.html
plt.figure(figsize=(12,4), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df, hue="level of education", palette="Accent")
# yanl yazarsan sana neri sunar
```

```python
# Palette=
# Choosing a palette from Matplotlib's cmap: https://matplotlib.org/tutorials/colors/colormaps.html
plt.figure(figsize=(12,4), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df, hue="level of education", palette="Dark2")
```

```python
# size="salary" --- az maa kk nokta, ok maa byk nokta
plt.figure(figsize=(12,4), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df, size="salary", hue="level of education", palette="Dark2")
```

```python
# size="salary" --- az maa kk nokta, ok maa byk nokta
plt.figure(figsize=(12,4), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df, size="work experience", hue="level of education", palette="Dark2")
```

```python
# s=200 --- tm noktalar 200 yap
plt.figure(figsize=(12,4), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df, s=200, hue="level of education", palette="Dark2")
```

```python
# alpha=0, alpha=1, 0=tam beyaz, 1=tam seilen renkte
plt.figure(figsize=(12,4), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df, size="salary", hue="level of education", palette="Dark2", alpha=0)
```

```python
# alpha=0, alpha=1, 0=tam beyaz, 1=tam seilen renkte
plt.figure(figsize=(12,4), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df, size="salary", hue="level of education", palette="Dark2", alpha=1)
```

```python
# alpha=0, alpha=1, 0=tam beyaz, 1=tam seilen renkte
plt.figure(figsize=(12,4), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df, size="salary", hue="level of education", palette="Dark2", alpha=0.3)
```

```python
# style="level of education"
plt.figure(figsize=(12,4), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df, size="salary", hue="level of education", palette="Dark2", alpha=1, style="level of education")
```
