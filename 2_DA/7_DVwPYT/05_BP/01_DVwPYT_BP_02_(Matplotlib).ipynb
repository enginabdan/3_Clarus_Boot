```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# In this lesson you will learn;
    # axes, canvas, figure concepts,
    # Functional, object-oriented method,
    # xlabel, ylabel, title to the plot,
    # Creating a figure object and adding axes to it.

fig = plt.figure()
# LARGE AXES
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Power of 4")
# SMALL AXES
axes2 = fig.add_axes([1,1,2,2])
axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title("Zoomed in")
plt.show()


fig = plt.figure(figsize=(2,2), dpi=200)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig("new_figure.png", bbox_inches = "tight")
```
