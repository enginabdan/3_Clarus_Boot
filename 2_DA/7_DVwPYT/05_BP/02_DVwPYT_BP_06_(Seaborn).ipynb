```python
plt.figure(dpi=100)
sns.heatmap(df.drop("Life expectancy", axis=1), linewidths=0.5, annot=True)
```

```python
plt.figure(dpi=100)
sns.heatmap(df.drop("Life expectancy", axis=1), linewidths=0.5, annot=True, cmap="viridis")
```

### CLUSTERMAP

```python
plt.figure(dpi=100)
sns.clustermap(df.drop("Life expectancy", axis=1), linewidths=0.5, annot=True, cmap="viridis", col_cluster=False)
```
