```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label="X vs X")
ax.plot(x, x**2, label="X vs X**2")
ax.legend(loc="lower left")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color="#D3DA2B",label="X vs X",lw=10,ls="--")  # RGB HEX CODE
ax.plot(x,x+1,color="purple",label="X vs X+1",linewidth=3,linestyle="dotted")
ax.legend()
# supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,x,color="#D3DA2B",label="X vs X",lw=10,ls="--")  # RGB HEX CODE
lines = ax.plot(x,x+1,color="purple",linewidth=5)
lines[0].set_dashes([5,2,5,2]) # 5 birim çizgi, 2 birim boşluk, 5 birim çizgi, 2 birim boşluk
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,x,color="#D3DA2B",label="X vs X",lw=10,ls="--")  # RGB HEX CODE
lines = ax.plot(x,x+1,color="purple",linewidth=5)
lines[0].set_dashes([1,2,1,2,10,2]) # 1 nokta, 2 boş, 1 nokta, 2 boş, 10 çizgi, 2 boş
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,x,color="#D3DA2B",label="X vs X",lw=10,ls="--")  # RGB HEX CODE
ax.plot(x,x,color="purple",linewidth=2,marker="o",ls="--",ms=10)
ax.plot(x,x+1,color="red",linewidth=2,marker="+",ls="--",markersize=20)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,x,color="#D3DA2B",label="X vs X",lw=10,ls="--")  # RGB HEX CODE
ax.plot(x,x,color="purple",linewidth=2,marker="s",ls="--",ms=10,markerfacecolor="red",markeredgewidth=4,markeredgecolor="orange")
```

```python
x = np.linspace(0, 5, 11)
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x+13, color="green", lw=3, ls='-', marker='o', markersize=10)
```

```python
x = np.linspace(0, 5, 11)
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x+13, color="green", lw=3, ls='-', marker='o', markersize=10)
ax.set_xlim([0, 3])
ax.set_ylim([13, 16])
```

```python
x = np.linspace(0, 5, 11)
fig, axes = plt.subplots()    
axes.plot(x, x**2, label="square")
axes.plot(x, x**3, label="cube")
axes.legend()
fig.savefig("square_cube.png")
```
