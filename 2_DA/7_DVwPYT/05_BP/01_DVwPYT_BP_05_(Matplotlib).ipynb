```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,x,color="#D3DA2B",label="X vs X",lw=10,ls="--")  # RGB HEX CODE
ax.plot(x,x,color="purple",linewidth=2,marker="s",ls="--",ms=10,markerfacecolor="red",markeredgewidth=4,markeredgecolor="orange")
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 4))
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[1].plot(x, y)
axes[1].set_title("title")
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
fig, axes = plt.subplots(2, 2, figsize=(8, 4))
axes[0,1].plot(x, y)
axes[0,1].set_title("title")
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x, y, color="red", lw=3, ls='-', marker='o', markersize=10)
```

```python
x = np.linspace(0, 5, 11)
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x+13, color="green", lw=3, ls='-', marker='o', markersize=10)
```

```python
x = np.linspace(0, 5, 11)
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x+13, color="green", lw=3, ls='-', marker='o', markersize=10)
ax.set_xlim([0, 3])
ax.set_ylim([13, 16])
```

```python
x = np.linspace(0, 5, 11)
fig, axes = plt.subplots()    
axes.plot(x, x**2, label="square")
axes.plot(x, x**3, label="cube")
axes.legend()
fig.savefig("square_cube.png")
```
