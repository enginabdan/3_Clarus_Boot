```python
# SUMMARY

# Merging,
# Joining,
# concat()

one = pd.DataFrame(data_one)
two = pd.DataFrame(data_two)

pd.concat([one, two], axis=1)
pd.concat([two, one], axis=1)

pd.concat([one, two], axis=0)
pd.concat([two, one], axis=0)

two.columns = one.columns # TWO Column names the same now with ONE

pd.concat([one, two], axis=0)

mydf = pd.concat([one, two], axis=0)
mydf.index = range(len(mydf))

registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})

pd.merge(registrations, logins, how="left", on="name") # Merge on registrations on left side
pd.merge(registrations, logins, how="right", on="name") # Merge on logins on right side
pd.merge(logins, registrations, how="left", on="name") # Merge on logins on left side
pd.merge(logins, registrations, how="right", on="name") # Merge on registrations on right side

pd.merge(left=registrations, right=logins, how="left", on="name") # Merge on registrations on left side
pd.merge(left=logins, right=registrations, how="left", on="name") # Merge on logins on left side
pd.merge(left=registrations, right=logins, how="right", on="name") # Merge on logins on right side
pd.merge(left=logins, right=registrations, how="right", on="name") # Merge on registrations on right side

pd.merge(left=logins, right=registrations, how="right") # Because the "name" the same

registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})

pd.merge(registrations, logins, how="outer", on="name") # Default on all left dataframe, the rest
pd.merge(logins, registrations, how="outer", on="name")

registrations = registrations.set_index("name")

pd.merge(registrations, logins, left_index=True, right_on="name", how="inner")

registrations = registrations.reset_index()
registrations.columns = ["reg_name", "reg_id"]

result = pd.merge(registrations, logins, how="inner", left_on="reg_name", right_on="name")

result.drop("reg_name", axis=1)
result.drop("name", axis=1)

registrations.columns = ["name", "id"]
logins.columns = ["id", "name"]

pd.merge(registrations, logins, how="inner", on="name")
pd.merge(registrations, logins, how="inner", on="name", suffixes=("_reg", "_log"))

pd.concat([data1,data2], axis=0)
```
