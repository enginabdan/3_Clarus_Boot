```python
import numpy as np
import pandas as pd
import warnings
```

```python
df = pd.read_csv("mpg.csv")
```

```python
df.head()
```

```python
# Hangi yıllara ait var
df["model_year"].unique()
```

```python
# Hangi yıllardan kaç adet var
df["model_year"].value_counts()
```

```python
# Model yıllarına göre grupla
df.groupby("model_year")
```

```python
# Model yıllarına göre grupla ve her sütunda mean bul
df.groupby("model_year").mean().head()
```

```python
# Ortalama mpg her model yılı için
df.groupby("model_year").mean()["mpg"].head()
```

```python
# model_year, cylinders göre grupla ve tüm verilerin mean bul
df.groupby(["model_year", "cylinders"]).mean().head()
```

```python
# Hangi columlara uygulanmış gör,
# model_year ve cylinders artık colum değil dikkat edilirse, index oldular
df.groupby(["model_year", "cylinders"]).mean().columns
```

```python
# Tüm index ihtimallerini gösteriyor
df.groupby(["model_year", "cylinders"]).mean().index
```

```python
# model_year ve cylinders göre indexle ve mpg nin meanini bul
df.groupby(["model_year", "cylinders"]).mean()["mpg"].head()
```

```python
# model_year göre grupladı ve istatistik bilgilerini her sütun için çıkardı
df.groupby("model_year").describe().transpose().head(15)
```

```python
# model_year ve cylinders göre indexleyip tüm sütunların meanlerini bulduk
year_cyl = df.groupby(["model_year", "cylinders"]).mean()
```

```python
year_cyl.head()
```

```python
# year_cyl gruplamada indexlerin isimlerini yazdırır
year_cyl.index.names
```

```python
# year_cyl gruplamada indexlerin hangi levellarda yapıldığını gösterir
year_cyl.index.levels
```

```python
# 70 model olanları getir
year_cyl.loc[70]
```

```python
# 70 ve 82 olanları getir
year_cyl.loc[[70,82]]
```

```python
# 70 model 4 silindirli araçları gösterir
year_cyl.loc[(70, 4)]
```

```python
# model_year leveldaki 70 olan key levelını getir
# Burada 2 tane key yazamayız bunu .loc[] ile yapabilirsin
year_cyl.xs(key=70, level="model_year")
```

```python
year_cyl.loc[[70,80]]
```

```python
# 4 cylinders olanları getirme 
four_cyl = year_cyl.xs(key=4, level="cylinders")
```

```python
four_cyl
```

```python
# önce "df[df["cylinders"].isin([6,8])]" ile filtrele,
# sonra "groupby(["model_year", "cylinders"])" ile grupla
# sonra da mean al buna göre
df[df["cylinders"].isin([6,8])].groupby(["model_year", "cylinders"]).mean().head(10)
```

```python
year_cyl.head()
```

```python
# Index Levelları swap yapabilirsin
year_cyl.swaplevel().head()
```

```python
# İndexteki dilediğin levelda asc veya dsc yapabilirsin
year_cyl.sort_index(level="model_year", ascending=False).head(10)
```

```python
df.head()
```

```python
# Tüm columlar için std ve mean göster
df.agg(["std", "mean"])
```

```python
# Sadece "mpg" için std ve mean göster
df.agg(["std", "mean"])["mpg"]
```

```python
# Her columda farklı istatistik talebinde kullanılır
# hesap yapılmayan alanlar NaN olur
df.agg({"mpg":["max", "mean"], "weight":["mean", "std"]})
```
