```python
# TASK: What are the top 5 most common country codes in the dataset?
```

```python
hotels['country'].value_counts()[:5]
```

```python
# TASK: What is the name of the person who paid the highest ADR (average daily rate)?
# How much was their ADR?
```

```python
hotels.sort_values('adr',ascending=False)[['adr','name']].iloc[0]
```

```python
# TASK: The adr is the average daily rate for a person's stay at the hotel.
# What is the mean adr across all the hotel stays in the dataset?
```

```python
round(hotels['adr'].mean(),2)
```

```python
# TASK: What is the average (mean) number of nights for a stay across the entire data set?
# Feel free to round this to 2 decimal points.
```

```python
hotels['total_stay_days'] = hotels['stays_in_week_nights'] + hotels['stays_in_weekend_nights']
round(hotels['total_stay_days'].mean(),2)
```

```python
# TASK: What is the average total cost for a stay in the dataset?
# Not average daily cost, but total stay cost. Feel free to round this to 2 decimal points.
```

```python
hotels['total_paid'] = hotels['adr'] * hotels['total_stay_days']
round(hotels['total_paid'].mean(),2)
```

```python
# TASK: What are the names and emails of people who made 5 "Special Requests"?
```

```python
hotels[hotels['total_of_special_requests'] == 5][['name','email']]
```

```python
# TASK: What percentage of hotel stays were classified as "repeat guests"?
# (Do not base this off the name of the person, but instead of the is_repeated_guest column)
```

```python
round(100 * sum(hotels['is_repeated_guest'] == 1) / len(hotels),2)
```

```python
# TASK: What are the top 5 most common last name in the dataset?
# Bonus: Can you figure this out in one line of pandas code?
# (For simplicity treat the a title such as MD as a last name,
# for example Caroline Conley MD can be said to have the last name MD)
```

```python
hotels['name'].apply(lambda name: name.split()[1]).value_counts()[:5]
```

```python
# TASK: What are the names of the people who had booked the most number children and babies for their stay?
# (Don't worry if they canceled, only consider number of people reported at the time of their reservation)
```

```python
hotels['total_kids'] = hotels['babies'] + hotels['children']
hotels.sort_values('total_kids',ascending=False)[['name','adults','total_kids','babies','children']][:3]
```

```python
# TASK: What are the top 3 most common area code in the phone numbers? (Area code is first 3 digits)
```

```python
print('Code - Total Count')
hotels['phone-number'].apply(lambda num:num[:3]).value_counts()[:3]
```

```python
# TASK: How many arrivals took place between the 1st and the 15th of the month (inclusive of 1 and 15) ?
# Bonus: Can you do this in one line of pandas code?
```

```python
hotels['arrival_date_day_of_month'].apply(lambda day: day in range(1,16)).sum()
```

```python
# HARD BONUS TASK: Create a table for counts for each day of the week that people arrived.
# (E.g. 5000 arrivals were on a Monday, 3000 were on a Tuesday, etc..)
```

```python
def convert(day,month,year):
    return f'{day}-{month}-{year}'
```

```python
hotels['date']  = np.vectorize(convert)(hotels['arrival_date_day_of_month'],
                                        hotels['arrival_date_month'],
                                        hotels['arrival_date_year'])
```

```python
hotels['date'] = pd.to_datetime(hotels['date'])
```

```python
hotels['date'].dt.day_name().value_counts()
```
