```python
outside = ["M1","M1","M1","M2","M2","M2"]
inside = [1,2,3,1,2,3]
multi_index = list(zip(outside, inside))
multi_index
```

```python
df5  = pd.DataFrame(np.random.randn(6,2), index=hier_index, columns=["A","B "])
df5
```
