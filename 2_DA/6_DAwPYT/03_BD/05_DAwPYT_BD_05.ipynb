#### groupby()

```python
df4.groupby("groups").mean()
```

#### aggregate([min, max, mean, ...])
#### agg({"var1":"min", "var2":"max"})

```python
df4.groupby("groups").aggregate([np.min, np.max, np.median])
```

```python
df4.groupby("groups").aggregate([np.min, np.max, np.median]).transpose()
```

```python
df4.groupby("groups").agg([np.min, np.max, np.median]).transpose()
```

```python
df4.groupby("groups").agg({"var1":"min", "var2":"max"})
```
