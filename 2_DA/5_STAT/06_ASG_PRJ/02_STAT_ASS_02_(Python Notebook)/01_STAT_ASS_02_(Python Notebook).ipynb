```python
import numpy as np
import pandas as pd
import scipy.stats
from scipy import stats
```

```python
samples = np.random.normal(100, 15, 1000) # (mean, sd, values)
print(samples)
```

```python
mean = np.mean(samples)
median = np.median(samples)
mode = stats.mode(samples)
print("mean: ", mean)
print("median: ", median)
print("mode: ", mode)
```

```python
min = np.min(samples)
max = np.max(samples)
q1 = np.percentile(samples, 25)
q3 = np.percentile(samples, 75)
iqr = q3 - q1
print("min: ", min)
print("max: ", max)
print("q3: ", q3)
print("q1: ", q1)
print("iqr (q3-q1): ", iqr)
```

```python
variance = np.var(samples)
std_dev = np.std(samples)
print("variance: ", variance)
print("std_dev: ", std_dev)
```

```python
skewness = stats.skew(samples) # (if "-" then left skew, "+" then right skew)
kurtosis = stats.kurtosis(samples)
print("skewness: ", skewness)
print("kurtosis: ", kurtosis)
```

```python
x = pd.Series(range(10, 20))
y = pd.Series([2, 1, 4, 5, 8, 12, 18, 25, 96, 48])
print(x)
print(y)
```

```python
r = scipy.stats.pearsonr(x, y)
print(r)
```

```python
rho = scipy.stats.spearmanr(x, y)
print("rho: ", rho)
```
