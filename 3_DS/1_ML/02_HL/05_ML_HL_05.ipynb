### Modelling

### Hopkins Test 
#### Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express no clusterability, a score tending to 1 express an uniform distrubution so clustering can't be useful and  0 express  not an uniform distrubution hence clusterering can be useful.

```python
from pyclustertend import hopkins
```

```python
hopkins(X, X.shape[0])
```

### Clustering with K-means

### Elbow Method

```python
ssd = []
K = range(2,10)

for k in K:
    model = KMeans(n_clusters=k)
    model.fit(X)
    ssd.append(model.inertia_)
```

```python
ssd
```

```python
sns.lineplot(x=K, y = ssd,markers="x")

plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
kmeans = KMeans()
visu = KElbowVisualizer(kmeans, k = (2,10))
visu.fit(X)
visu.poof()
```

### Model Building and label visualisation

```python
model = KMeans(n_clusters=4)
model.fit(X)
```

```python
model.inertia_
```

```python
model.labels_
```

```python
value_df["cluster"] = model.labels_
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(value_df["Airforce Strength_Aircraft Strength value"], 
                value_df["Navy Strength_Navy Fleet Strengths value"],
                hue= value_df.cluster,
                palette= "viridis");
```

```python
fig = px.scatter_3d(x=value_df["Airforce Strength_Aircraft Strength value"],
                    y=value_df["Finances_defense spending budget value"],
                    z=value_df["Finances_External Debt value"],
                    color=value_df.cluster,
                    labels = {"x": "Airforce Strength_Aircraft Strength value",
                              "y": "Finances_defense spending budget value",
                              "z": "Finances_External Debt value"},
                    hover_name=value_df["country"])

fig.show()
```

```python
centroids = model.cluster_centers_

plt.scatter(X[:,18], X[:,19], c = value_df.cluster, s = 50, cmap = "viridis")
plt.scatter(centroids[:,18], centroids[:,19], c = "black", s = 200, alpha=0.5);
```

### Evaluate 

```python
from sklearn.metrics import silhouette_score
```

```python
# towards to 1 means, data good clustered by model
print(f'Silhouette Score(n=4): {silhouette_score(X, value_df.cluster)}')
```

```python
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

```python
den = pd.DataFrame(X, columns=value_df.drop(["country","cluster"], axis=1).columns)
den["cluster"] = value_df["cluster"]

clus_pred = den.groupby("cluster").mean().T
clus_pred
```

```python
fig= px.line(clus_pred)
fig.show()
```

### Hierarchical Clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

### Dendrogram

```python
hc_ward = linkage(y = X, method = "ward")
hc_complete = linkage(X, "complete")
hc_average = linkage(X, "average")
hc_single = linkage(X, "single")
```

```python
plt.figure(figsize = (20,12))

plt.subplot(221)
plt.title("Ward")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_ward, leaf_font_size = 10)

plt.subplot(222)
plt.title("Complete")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_complete, leaf_font_size = 10)

plt.subplot(223)
plt.title("Average")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_average, leaf_font_size = 10)

plt.subplot(224)
plt.title("Single")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_single, leaf_font_size = 10);
```
